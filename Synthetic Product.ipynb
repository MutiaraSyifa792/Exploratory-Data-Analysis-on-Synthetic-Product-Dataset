```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/archive (1).zip')

# Display the first few rows
df.head()
```

```python
# Overview of dataset
df.info()
```

```python
# Count missing values per column
df.isnull().sum()
```

```python
duplicate_count = df.duplicated().sum()
print("\nTotal Duplicate Rows:", duplicate_count)
```

```python
# Example strategy: drop rows with missing values
df = df.dropna()
```

```python
# Check for duplicates
df.duplicated().sum()

# Drop duplicates
df = df.drop_duplicates()
```

```python
# Summary statistics
df.describe()
```

```python
# Histogram of Price
sns.histplot(df['Price'], kde=True)
plt.title('Distribution of Price')
plt.show()

# Boxplot of Ratings
sns.boxplot(x=df['Rating'])
plt.title('Boxplot of Ratings')
plt.show()

# Countplot of Stock status
sns.countplot(x='Stock', data=df)
plt.title('Stock Availability')
plt.show()
```
